# HW1_3: 1. Matplotlib
ใช้ Matplotlib ในการสร้างเส้นโค้ง parabola แบบหงายหยาบ ๆ โดยมีรายละเอียดดังนี้

\
1.กำหนดจุด (x, y) จาก np.array สำหรับสร้างเส้น parabola แบบหงายอย่างง่าย\
2.ตั้งชื่อกราฟว่า "2 separated lines"\
3.เส้นซีกซ้ายเป็นสีแดง และเส้นซีกขวาเป็นสีน้ำเงิน

\
image.png

```python
# HW1_3: 1

import matplotlib.pyplot as plt
import numpy as np

# Create data
x = np.linspace(-4, 4, 8)
y = x**2

# Find vertex (minimum point)
vertex_idx = np.argmin(y)

# Plot left side (red) and right side (blue)
plt.plot(x[:vertex_idx+1], y[:vertex_idx+1], 'r-', linewidth=2)
plt.plot(x[vertex_idx:], y[vertex_idx:], 'b-', linewidth=2)

# Set title and limits
plt.title('2 separated lines')
plt.xlim(-5, 5)
plt.ylim(0, 10)

plt.show()
```

ใช้ Matplotlib ในการสร้างกราฟวงกลมโดยมีรายละเอียดดังนี้

\
1.กำหนดฟังก์ชั่น x, y จาก wave function ที่มีใน numpy library สำหรับการสร้างวงกลม 1 หน่วยย่างง่าย\
2.ตั้งชื่อกราฟว่า "2 separated circle lines"\
3.วงกลมซีกซ้ายเป็นสีเขียว และวงกลมซีกขวาเป็นสีชมพู

\
image.png

```python
# HW1_3: 1

import matplotlib.pyplot as plt
import numpy as np

# Create left semicircle (green)
theta_left = np.linspace(np.pi/2, 3*np.pi/2, 50)
x_left = np.cos(theta_left)
y_left = np.sin(theta_left)

# Create right semicircle (pink)
theta_right = np.linspace(-np.pi/2, np.pi/2, 50)
x_right = np.cos(theta_right)
y_right = np.sin(theta_right)

plt.plot(x_left, y_left, color='green', linewidth=2)
plt.plot(x_right, y_right, color='pink', linewidth=2)
plt.title('2 separated circle lines')
plt.axis('equal')
plt.show()
```
